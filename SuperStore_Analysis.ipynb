```python
gbsegment=df.groupby('Segment').sum()
```

```python
plt.figure(figsize=(14,5))
sns.set_style('whitegrid')

plt.subplot(1,2,1)
plt.title('Total Profit')
sns.barplot(x=gbsegment.index , y='Profit' , data=gbsegment , errorbar=None) 

plt.subplot(1,2,2)
plt.title('Total Sales')
sns.barplot(x=gbsegment.index , y='Sales' , data=gbsegment , errorbar=None) 
```
